```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
from sklearn import impute
```

### Imputation Example

```python
df = pd.read_csv("sc.csv")
df
```

```python
df.isnull().sum()
```

```python
df[df["L_LatVent"].isnull()]["L_LatVent"]
```

```python
df["L_LatVent"].fillna(df["L_LatVent"].mean(), inplace = True)
df.isnull().sum()
```

```python
model = impute.SimpleImputer(missing_values=np.nan, strategy="mean")

imp_df = pd.DataFrame( model.fit_transform( df ) )
imp_df.columns = df.columns
#imp_df.index = df.index
imp_df.isnull().sum()
```
