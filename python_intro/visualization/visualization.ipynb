```python
x = [1,2,3,4]
y = [1,4,9,16]
brent={"x":x, "y":y }
plt.plot( "x", "y", "bs--",data=brent )
plt.xlabel( "x ({}) axis".format(5) )
plt.ylabel( "y axis" )
plt.show()
```

```python
x = [1,2,3,4]
y = [1,4,9,16]
plt.stem( x, y, "r--", markerfmt="go", basefmt="k" )
plt.show()
```

```python
x = np.array( [1, 2, 3, 4] ) # works with Numpy arrays too!!
y = np.array( [1, 4, 9, 16] )
plt.plot( x, y )
plt.ylabel("some numbers")
plt.show()
```

```python
matrix = np.random.rand(10,2)
plt.plot( matrix[:,0], matrix[:,1], "bo" )
plt.title( "Matrix example")
plt.show()
```

#### Line formating

For every x, y pair of arguments, there is an optional third argument which is the format string that indicates the color and line type of the plot. The default format string is 'b-', which is a solid blue line.



```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], "ro-")
plt.axis([0, 6, 0, 20])
plt.show()
```

```python
# evenly sampled intervals
t = np.arange(0., 5., 0.2)

# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
#plt.plot(t, t, 'r--')
#plt.plot(t, t**2, 'bs')
#plt.plot(t, t**3, 'g^')
plt.legend(["red", "blue", "green"])
plt.show()
```

```python
plt.scatter( x, y, s=50, c="m", marker="x" )
plt.title( "Scatter - List Data")
plt.show()
```

```python
plt.scatter( matrix[:,0], matrix[:,1], s=25, c='blue' )
plt.title( "Scatter - Matrix Data")
plt.show()
```

### Plotting with Dictionaries

```python
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

### Categorial Data Plots

Documentation <a href="https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplot.html?highlight=subplot#matplotlib.pyplot.subplot">subplot</a>

```python
names = ["Group A", "Group B", "Group C"]
values = [1, 10, 100]
plt.bar(names, values, color=["blue", "red", "orange"] )
plt.show()
```

```python
plt.figure(figsize=(9, 3))

plt.subplot(131)
plt.bar(names, values)
plt.title("test")
plt.subplot(132)
plt.scatter(names, values)
plt.subplot(133)
plt.plot(names, values)
plt.suptitle('Categorical Plotting')
plt.show()
```

### Controlling Line Properties


```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure()
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k', linewidth=2.0)

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--', linewidth=5.0 )
plt.show()
```

### Mathematical Annotations ###


```python
mu, sigma = 115, 15
x = mu + sigma * np.random.randn(10000)

# the histogram of the data
plt.figure(figsize=(10, 5))
plt.hist(x, 50, facecolor="b", alpha=0.75)
plt.xlabel("Length [cm]")
plt.ylabel("Probability")
plt.title("Aardvark lengths\n (not really)")
plt.text(80, 500, "$\mu=115,\ \sigma=15$")
plt.grid(True)
```

### Text Annotations ###


```python
t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)

plt.figure()
plt.plot(t, s, linewidth=2)
plt.annotate("local max", xy=(2, 1), xytext=(3, 1.5),
            arrowprops=dict(facecolor='black', shrink=0.05) )

plt.ylim(-2, 2)
plt.show()
```
